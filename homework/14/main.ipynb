```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# task 1
weeks = ['Тиждень 1', 'Тиждень 2', 'Тиждень 3', 'Тиждень 4']

traffic = [1500, 1600, 1700, 1800]

sales_product_a = [300, 350, 400, 450]
sales_product_b = [200, 250, 300, 350]
sales_product_c = [100, 150, 200, 250]


plt.plot(weeks, traffic, label="Трафік", color="blue", marker="o", linestyle="-")
plt.plot(weeks, sales_product_a, label="Продажі A", color="red", marker="s", linestyle="-")
plt.plot(weeks, sales_product_b, label="Продажі B", color="green", marker="^", linestyle="-")
plt.plot(weeks, sales_product_c, label="Продажі C", color="orange", marker="d", linestyle="-")

plt.title("Веб трафік")
plt.xlabel("Тиждень")
plt.ylabel("Кількість")

plt.legend()
plt.grid(True)

plt.show()
```

```python
# task 2

x = np.linspace(0, 10, 1000)
y = np.sin(x) * np.cos(x)


plt.figure(figsize=(12, 6))

plt.plot(x, y, color="purple", label="y = sin(x) * cos(x)")

plt.title("Графік функції")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.tight_layout()

plt.show()
```

```python
# task 3

sizes = [50, 25, 15, 10]
labels = ['Продажі', 'Інвестиції', 'Ліцензії', 'Інше']
colors = ["red", "blue", "yellow", "purple"]
plt.pie(sizes, labels=labels, colors=colors)
plt.title("Кольори сегментів")
plt.show()
```
