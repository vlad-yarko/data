```python
# Task 2

 
data = httpx.get("https://jsonplaceholder.typicode.com/users").json()


df = pd.DataFrame(data)
df.to_json("raw.json", orient="records", indent=2)


required = ["id", "name", "email"]
df["missing"] = df.apply(lambda row: [f for f in required if pd.isna(row.get(f)) or row.get(f) == ""], axis=1)
 
valid = df[df["missing"].map(len) == 0]
invalid = df[df["missing"].map(len) > 0]
 
print(valid[required])
print("Невалідні:", invalid[["id", "missing"]])
 

```
