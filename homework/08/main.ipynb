```python
import numpy as np
```

```python
array = np.array([12, -5, 7, 9, -3, 15])
array = np.append(array, 10)
array
```

```python
array = np.array([
    [5, 2, 8],
    [1, 7, 4],
    [3, 6, 9]
])
array = np.delete(array, 1, axis=0)
array
```

```python
conversions = np.array([50, 65, 80, 45, 70, 90, 55, 85, 60, 75])
print(conversions)

print(np.median(conversions))
print(np.var(conversions))
print(np.std(conversions))

array1 = np.random.rand(5)
array2 = np.random.randint(1, 11, size=5)
array3 = np.random.uniform(5, 15, size=5)
array4 = np.random.normal(0, 1, size=5)

combined_conversions = np.concatenate([conversions, array1, array2, array3, array4])
print(combined_conversions)

mean_of_combined_conversions = combined_conversions.mean()
print(mean_of_combined_conversions)
print(mean_of_combined_conversions > 2000)
```
