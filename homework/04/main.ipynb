```python
import numpy as np
```

```python
array = np.arange(10, 20)
# We can call or general function of numpy, or methods of arrays directly
print(np.sum(array))
print(array.sum())
print(np.mean(array))
print(array.mean())
print(np.min(array))
print(array.min())
print(np.max(array))
print(array.max())
```

```python
array = np.random.rand(1000).astype(np.float64)
print(np.sum(array))
print(array.sum())
print(np.mean(array))
print(array.mean())
print(np.min(array))
print(array.min())
print(np.max(array))
print(array.max())
```

```python
array = np.ones((5, 5), dtype=int)
print(array[:, 1])
print(array[1])
print(array.flatten())
```

```python
array = np.random.rand(500000)
print(np.sum(array))
print(array.sum())
print(np.mean(array))
print(array.mean())
print(np.min(array))
print(array.min())
print(np.max(array))
print(array.max())
```
