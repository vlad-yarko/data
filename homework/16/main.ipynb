```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Task 1


df = sns.load_dataset('tips')
sns.displot(
    data=df,
    x='total_bill',
    hue='smoker',
    bins=20,
    kde=True
)
plt.title('     ')
plt.xlabel('  (USD)')
plt.ylabel(' ')
plt.show()


```

```python
# Task 2

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')
sns.histplot(data=df, x='total_bill', bins=20, kde=True, color='brown', alpha=0.7, line_kws={'linewidth': 3})
plt.title('     ')
plt.xlabel('  (USD)')
plt.ylabel(' ')
plt.show()
```

```python
# Task 3

df = sns.load_dataset('penguins')

df = df.dropna()

df_encoded = pd.get_dummies(df, drop_first=True)

corr_matrix = df_encoded.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('')
plt.show()
```
