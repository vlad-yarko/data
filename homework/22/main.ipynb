```python
import pandas as pd
import time
```

```python
file_path = "homework.csv"
```

```python
# Task 1
#NOTE: this code would be useful if dataset had "Price" column, but it doesn't
# So I just add price column to each chunk with default value


chunk_size = 10000
total_sum = 0
total_count = 0


for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    #BUG: dataset does not have price column
    chunk["Price"] = 0
    total_sum += chunk["Price"].sum()
    total_count += chunk["Price"].count()


average_price = total_sum / total_count
average_price
```

```python
# Task 2


start = time.time()

df = pd.read_csv(file_path)

print("Original read time:", time.time() - start)
print("Original memory:", df.memory_usage(deep=True).sum())


dtype_opt = {
    "id": "int32",
    "age": "int8",
    "score": "float32"
}

start = time.time()

df_opt = pd.read_csv(file_path, dtype=dtype_opt)

print("Optimized read time:", time.time() - start)
print("Optimized memory:", df_opt.memory_usage(deep=True).sum())
```

```python
df = pd.read_csv(file_path)
df.to_parquet("homework.parquet")


start = time.time()
pd.read_csv("homework.csv")
csv_time = time.time() - start


start = time.time()
pd.read_parquet("homework.parquet")
parquet_time = time.time() - start


print("CSV read time:", csv_time)
print("Parquet read time:", parquet_time)
```
