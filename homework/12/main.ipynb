```python
import pandas as pd
```

```python
# Task 1

data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet'],
    'Price': [1200, 800, 400],
    'Stock': [50, 150, 200]
}
df = pd.DataFrame(data)

new_row = {'Product': 'USB Hub', 'Price': 40, 'Stock': 150, 'Warranty': '2 years'}

df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

df
```
