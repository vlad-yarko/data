```python
import folium

location = [50.4501, 30.5234]

m = folium.Map(location=location, zoom_start=10)

print(m)

folium.Marker(
    location=[50.4501, 30.5234],
    popup='Центр Києва',
    icon=folium.Icon(icon='info-sign')
).add_to(m)

print(m)

polygon = [
    [50.4547, 30.5238],
    [50.4547, 30.5338],
    [50.4447, 30.5338],
    [50.4447, 30.5238],
    [50.4547, 30.5238]
]

folium.Polygon(
    locations=polygon,
    color='blue',
    fill=True,
    fill_color='cyan'
).add_to(m)

print(m)
```
